### CountVectorizer Concept:

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()

data = ["Love Data science", "Love Machine Learning"]

vect_data = vect.fit_transform(data)

print(vect.get_feature_names_out()) #Bag of words

print("")

print(vect_data.toarray())
```

Bag of Words (BoW): A technique that converts text into numerical vectors based on word frequency. Doesn't pay attention on ordering of words. CountVectorizer: A Scikit-learn tool that implements the Bag of Words model to convert text into a matrix of word counts.

### First, we've to convert it into array then only it'll be converted into numbers

```python
x = np.array(df["Text"])
y = np.array(df["language"])
```

```python
x
```

```python
y
```

```python
cv = CountVectorizer()
X = cv.fit_transform(x) 
#we only want to tranform input feature (Text) into number. no need of transforming language data into number.
```

### Train Test Split:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.15, random_state=42)
```

```python
X_train
```

```python
print(X_train)
```

```python
print(y_test)
```

### Naive Bayes Algorithm:

A probability-based classification algorithm that uses Bayes’ Theorem and assumes that all features are independent of each other.Example:
Used in spam email detection, sentiment analysis, and text classification.

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
user = input("Enter a text")

data = cv.transform([user]).toarray()
output = model.predict(data)
print(output)
```
